Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_boston

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
boston = load_boston()
boston.keys()
```

```python
from sklearn.model_selection import train_test_split
# Данные о недвижимости ( с признаками об объектах недвижимости )
data = boston["data"]
# Названия колонок этих признаков
feature_names = boston["feature_names"]
# Целевые значения цен на недвижимость
target = boston["target"]
# Датафрейм с признаками по схеме
X = pd.DataFrame(data, columns=feature_names)
# Датафрейм уже с целевым признаком цены на недвижимость
y = pd.DataFrame(target, columns=["price"])
# Поделим на тестовые и тренировочные данные
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
# Модель линейной регресии
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Обучим на тренировочных данных
lr.fit(X_train, y_train)
```

```python
# Предскажем на тестовых данных
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
# Сопоставим реальные и предсказанные
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head()
```

```python
import seaborn as sns
# Сопоставим визуально реально и предсказанные значения
# sns.displot(data=(check_test["y_test"],check_test["y_pred"]), legend=True, kde=True)
sns.jointplot(data=(check_test["y_test"],check_test["y_pred"]), kind="kde")
```

Чисто визуально видны различия, есть необходимость оченить метриками

```python
# Оценим модель R2 коэффициентом детерминации
from sklearn.metrics import r2_score

r2_linear = r2_score(check_test["y_test"], check_test["y_pred"])
r2_linear
```

Чем Ближе R2 к 1, тем лучше
